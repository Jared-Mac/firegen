```python
 
import tensorflow as tf
import numpy as np
import math
import matplotlib.pyplot as plt
import tensorflow as tf
# import tensorflow_datasets as tfds
from tensorflow import keras
from keras import layers
import io
import imageio

import models 
import generators 
import utils 
from setup import *
```

```python
model = models.DiffusionModel(image_size, 1, 3, widths, block_depth)
```

```python
optimizer=tf.keras.optimizers.legacy.Adam
model.compile(
    optimizer=optimizer(
        learning_rate=1e-5
    ),
    loss=keras.losses.mean_absolute_error,
)
```

```python
import pickle
data = None
with open("data.pkl", "rb") as f:
    data = pickle.load(f)
```

```python
data_gen = generators.FireDataGenerator(data)
```

```python
for x in range(np.shape(data)[0]):
    print(np.shape(data_gen.__getitem__(x)))
```

```python

def saver(epoch, logs):
    model.network.save_weights("weights/"+str(epoch)+"diffusion_addons")
    model.ema_network.save_weights("weights/"+str(epoch)+"diffusion_addons_ema")

reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor='i_loss', factor=0.5,
                              patience=2, min_lr=0)


# run training and plot generated images periodically
history = model.fit(
    data_gen,
    epochs=30,
    steps_per_epoch=14000,
    #validation_data = val_generator,
    batch_size=1,
    callbacks=[
        reduce_lr,
        #keras.callbacks.LambdaCallback(on_epoch_end=model.plotter),
        keras.callbacks.LambdaCallback(on_epoch_end=saver)
    ],
)

```
